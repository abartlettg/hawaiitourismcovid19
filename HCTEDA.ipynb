```python
# EDA for HCT Project

# Started in R, but may change over to Python or use both languages going forward

# Based on what I've tried in R already and searching the internet, it appears that plotly.express.timeline is the
# best function for creating the type of gantt chart I want to use for the Rules Timeline, but plotly.express in not
# included in the plotly package for R

# Because some data cleanup/manipulation has already been done to the hcrt (Hawaii COVID Rules Timeline) data, I have
# extracted the processed file from R (not the raw data file) and will read it in here.

import pandas as pd
import plotly as px 
import plotly.express as px
import plotly.graph_objects as go

hcrt = pd.read_csv('hcrt.csv')
 
# displaying the contents of the CSV file
hcrt
```

```python
# Create gantt chart w/plotly.express

px.timeline(hcrt, x_start="start_dt", x_end="end_dt", y="uid", color="type")
# This works well and will be simplified by categorizing types of Rules and levels of restriction
```
