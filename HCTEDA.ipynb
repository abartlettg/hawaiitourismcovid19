```python
# EDA for HCT Project

# Started in R, but may change over to Python or use both languages going forward

# Based on what I've tried in R already and searching the internet, it appears that plotly.express.timeline is the
# best function for creating the type of gantt chart I want to use for the Rules Timeline, but plotly.express in not
# included in the plotly package for R

# Because some data cleanup/manipulation has already been done to the hcrt (Hawaii COVID Rules Timeline) data, I have
# extracted the processed file from R (not the raw data file) and will read it in here.

import pandas as pd
#import plotly as px 
import plotly.express as px
import plotly.graph_objects as go

hcrt = pd.read_csv('hcrt.csv')
hcrt
```

```python
# Create gantt chart w/plotly.express

px.timeline(hcrt, x_start="start_dt", x_end="end_dt", y="uid", color="type")
# This works well and will be simplified by categorizing types of Rules and levels of restriction
```

```python
# Read in other data files to practice w/Plotly and see if this is what I want to use for all the data
# Files are output from R w/some preprocessing already done

ccbrs = pd.read_csv('ccbrs.csv')
tsdc = pd.read_csv('tsdc.csv')
```

```python
fig = px.histogram(ccbrs, x="DateAdded", color="Risk")
fig
# This is counting the number of records added, not summing the number of cases
```

```python
# First, Need to Aggregate ccbrs

ccbrstots = ccbrs.groupby(['DateAdded', 'Risk']).sum('TotNumber')
```

```python
ccbrstots = ccbrstots.drop(columns = 'Unnamed: 0') 
ccbrstots = ccbrstots.drop(columns = 'Count')
ccbrstots = ccbrstots.drop(columns = ['BigIslandCnt','OahuCnt','KauaiCnt','MauiCnt']) 
```

```python
ccbrstots.head(30)
```

```python
# Graphing COVID Cases By Resident Status Aggregated
```

```python
fig = px.line(ccbrstots, x=ccbrstots.DateAdded, y=ccbrstots.TotNumber, color=ccbrstots.Risk)
fig
# Having issues w/aggregating before using plotly.express
# Dplyr in R is easier to aggregate with
```
